```python
import os
import pandas as pd
import numpy as np

# Change this number to select a different batch
# Tip: Keep batch_config.yaml open while viewing this document to know which configurations are which
batch_num = 2
batch_base = os.path.join("..", "batches", f"batch{batch_num}")
config_folders = os.listdir(batch_base)


def average_data_from_config(config_folder):
    runs = os.listdir(config_folder)
    total_avg_time_in_class = 0
    total_avg_total_drinks = 0
    total_avg_time_in_queue = 0 
    for run in runs:
        data_file = os.path.join(config_folder, run, "data.csv")
        df = pd.read_csv(data_file)
        
        print(sum)

        total_avg_time_in_class += df.groupby(["Day", "Name"])["Time In Class"].max().groupby("Day").mean().mean()
        total_avg_total_drinks += df.groupby(["Day", "Name"])["Total Drinks"].max().groupby("Day").mean().mean()
        total_avg_time_in_queue += np.mean(df[df["Action"] == "Co"]["Duration"])

    total_avg_time_in_class = total_avg_time_in_class / len(runs)
    total_avg_total_drinks = total_avg_total_drinks / len(runs)
    total_avg_time_in_queue = total_avg_time_in_queue / len(runs)
    return [config_folder, total_avg_time_in_class, total_avg_total_drinks, total_avg_time_in_queue]

table = []
for config in config_folders:
    table.append(average_data_from_config(os.path.join(batch_base, config)))

pd.DataFrame(table, columns=["Config Folder", "Average time spent in class per day", "Average total drinks per day", "Average time in queue"])
```
